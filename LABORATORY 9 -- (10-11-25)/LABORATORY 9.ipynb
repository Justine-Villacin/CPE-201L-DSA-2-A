```python
# Queue implementation in python using Array

# Creating a queue
def create_queue():
    queue = []
    return queue

# Check if the queue is empty
def is_empty(queue):
    return len(queue) == 0

# Adding items into the queue (enqueue)
def enqueue(queue, item):
    queue.append(item)
    print("Enqueued Element: " + str(item))

# Removing an element from the queue (dequeue)
def dequeue(queue):
    if (is_empty(queue)):
        return "The queue is empty"
    return queue.pop(0)


print("=== ARRAY-BASED QUEUE IMPLEMENTATION ===\n")
queue = create_queue()
enqueue(queue, 1)
enqueue(queue, 2)
enqueue(queue, 3)
enqueue(queue, 4)
enqueue(queue, 5)

print("The elements in the queue are: " + str(queue))
```
