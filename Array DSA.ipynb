```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("\nTraversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr):
    reverse_array = arr[::-1]
    print(f"\n{reverse_array}")

def sort(arr):
    sorted_array = sorted(arr)
    print(f"\n{sorted_array}")

def insert(arr, element, index):
    arr.insert(index, element)
    return arr

def delete(arr, index):
    del arr[index]
    return arr

def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse Array")
        print("3. Sorted Array")
        print("4. Insert Element")
        print("5. Delete Element")
        print("6.Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse(arr)
        elif choice == '3':
            sort(arr)
        elif choice == '4':
            element = float(input("Enter the element to insert: "))
            index = int(input("Enter the index to insert at: "))
            arr = insert(arr, element, index)
            print("\nUpdated array:", arr)
        elif choice == '5':
            index = int(input("Enter the index to delete: "))
            arr = delete(arr, index)
            print("\nUpdated array:", arr)
        elif choice == '6':
            print("\nExiting the program.")
            break
        else:
          print("\nInvalid choice. Please enter a number between 1 to 6.")

if __name__ == "__main__":
    main()

```
